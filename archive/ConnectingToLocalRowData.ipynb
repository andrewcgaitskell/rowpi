```python
top = figure(width=300, height=300, y_axis_type='datetime')

top.yaxis.formatter = DatetimeTickFormatter(seconds=["%M:%S"],
                                            minutes=["%M:%S"],
                                            minsec=["%M:%S"],
                                            hours=["%M:%S"])
# a line works fine with time objects
top.line(xs, ys)

output_notebook()

show(top)
```
